```python
import pandas as pd
import pylab as pl
import numpy as np
import scipy.optimize as opt
from sklearn import preprocessing
import matplotlib.pyplot as plt
```

```python
#Load Data From CSV File
colon_df = pd.read_csv("colon_cancer.csv")
colon_df.head()
colon_df.tail()
```

```python
# list(data) or
col_names = list(colon_df.columns)
```

```python
#Lets select some features for the modeling. Also we change the target data type to be integer, as it is a requirement by the skitlearn algorithm:
colon_df = colon_df[col_names[1:2002]]
colon_df
```

```python
ax = colon_df[colon_df['Class'] == 'Abnormal'][0:62].plot(kind='scatter', x='H55933', y='R39465', color='DarkBlue', label='Abnormal');
colon_df[colon_df['Class'] == 'Normal'][0:62].plot(kind='scatter', x='H55933', y='R39465', color='Yellow', label='Normal', ax=ax);
plt.show()
```
